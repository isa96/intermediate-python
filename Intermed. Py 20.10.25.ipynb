#### 5. Get rid of the missing value:
#### 5.a. Drop columns which has missing value that cannot be filled with any logical value (like name, host_id, etc)

```python
# Mengecek feature mana yang memiliki missing value dan berapa jumlah missing valuenya.
data.isna().sum()
```

Terdapat 4 feature yang memiliki missing value, yaitu kolom 'name' = 16 missing value, 'host_name' = 21 missing value, 'last_review' = 10.052 missing value dan kolom 'reviews_per_month' = 10.052 missing value.\
Kolom 'name', 'host_id' dan 'host_name' akan di drop.

```python
# Mengedrop feature (kolom) 'name', 'host_id' dan 'host_name'
# untuk mengamankan dataframe 'data', maka data untuk di drop diberi nama 'data1'
data1 = data.copy()

data1.drop(['name', 'host_id', 'host_name'], axis = 1, inplace = True) 
```

```python
# Cek data 1 setelah di drop kolom 'name', 'host_id' dan 'host_name'
data1.isna().sum()
```

#### 5. Get rid of the missing value:
#### 5.b. Fill the missing value in last_review with the most common date

```python
# Cek 5 data pertama dari data1
data1.head()
```

```python
# Mengganti missing value pada kolom 'last_review' dengan tanggal yang sering muncul (most common date)
data1['last_review'] = data1['last_review'].fillna(data1['last_review'].mode()[0])
data1.isna().sum()
```

```python
# Cek 5 data pertama setelah di replace missing valuenya dengan tanggal yang sering muncul
data1.head()
```

#### 5. Get rid of the missing value:
#### 5.c.  Fill the missing value in reviews_per_month with the mean

```python
# Mengganti missing value pada kolom 'reviews_per_month' dengan nilai mean
data1['reviews_per_month'] = data1['reviews_per_month'].fillna(data1['reviews_per_month'].mean())
data1.isna().sum()
```

```python
# Cek hasil replace missing value pada data1 pada kolom 'reviews_per_month' dengan nilai mean
data1.head()
```

#### 6. Make category based on the price: Cheap, Medium, and Expensive

```python
# Memasukkan kategory harga pada kolom baru yang disebut 'price_cat', kriteri Cheap, Medium dan Expensive diambil
# dari analisa statistika di atas (mean, median etc)
# Cheap <69 --> Q1 (25%)
# Medium 69 - 175 --> Q1 - Q3
# Expensive >175 --> Q3 (75%)

data1["price_cat"] = np.where(data1["price"] <69, "Cheap", np.where(data1['price']<175, "Medium", "Expensive"))
```

```python
# Cek index kolom dari data1 setelah ditambah kolom 'price_cat'
data1.info()
```

#### 8. Visualize the price to compare the price of available rooms

```python
# Library yang digunakan
import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Plot menggunakan seaborn style
plt.style.use('seaborn')
```

```python
# Cek distribusi datanya dengan boxplot
plt.boxplot(x=data1['price'])
```

Dari cek distribusi data dengan boxplot, terlihat datanya tidak normal dimana data tergolong Menceng Kanan

```python
# Menggunakan Scatter plot untuk melihat dengat detail sebaran data price vs room availability_365
plt.scatter(data1['price'], data1['availability_365'])
plt.title('Relationship Between Price and Room Availability')
plt.xlabel('Price')
plt.ylabel('Room Availability')
plt.show()
```

Dari scatter plot di atas terlihat, mayoritas harga itu di range 0-1000, dan terdapat banyak room yang available

```python
# Karena data price banyak yang outlier, maka dicoba untuk melihat price antara 0-300 hubungannya 
# dengan available room

# Filter datanya untuk mengambil data price pada data1 yang nilainya dibawah 300 dan diberi nama data frame harga
harga = data1.loc[data1["price"]<300]
```

```python
# Visualisasi kembali dengan scatter plot data price
plt.scatter(harga['price'], harga['availability_365'])
plt.title('Relationship Between Price and Room Availability')
plt.xlabel('Price')
plt.ylabel('Room Availability')
plt.show()
```

```python
# Untuk distribusi data price pada data frame harga untuk membandingkannya dengan ketersediaan ruang menggunakan Histogram

harga.hist(column='price', bins=30,
           grid=True, figsize=(12,8), color='#86bf91', zorder=3, rwidth=0.9)

plt.title('Room Availability Based on Price')
plt.xlabel('Price (USD)')
plt.ylabel('Room Availability (Room)')

plt.tight_layout()
plt.show()
```

Dari visualisasi data di atas terdapat kesimpulan sebagai berikut :
1. Harga ruangan mayoritas di harga antara USD 0 - 1000
2. Ruang tersedia paling banyak di harga antara USD 60-70
