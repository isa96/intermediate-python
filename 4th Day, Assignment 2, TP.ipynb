### Exercise 5

1. Create a new column with the following condition:\
    a) >10    : Long Trip\
    b) 5-10   : Medium Trip\
    c) <5     : Short Trip

```python
import numpy as np
```

```python
# Terlebih dahulu list kriteria dalam object conditions
conditions = [
    (uber['MILES*'] <= 5),
    (uber['MILES*'] > 5) & (uber['MILES*'] <= 10),
    (uber['MILES*'] > 10)
    ]

# Kemudian masukkan nilai dari kriteria yang diinginkan
values = ['Short Trip', 'Medium Trip', 'Long Trip']

# Kemudian buat kolom baru dengan nama Distance dan masukkan kriteria ke dalam kolom baru
uber['DISTANCE'] = np.select(conditions, values)
uber
```
